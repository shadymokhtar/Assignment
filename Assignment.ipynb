```python
import pandas as pd
%matplotlib inline
import matplotlib 
import matplotlib.pyplot as plt

#Get both tables, bind them and remove 43 rows of messed up data 

feb=pd.read_excel("UK_feb_abc.xlsx")
mar=pd.read_excel("UK_Mar_abc.xlsx")
all_=feb.append(mar)
all_=all_[all_["Bin Price"]>0]
```

## What are the top selling main categories?

```python
plt.figure(figsize=(10,3))
plt.xticks(fontsize=12)
all_["Meta Category Name"].value_counts().plot(kind='bar')
plt.title("Top Selling Main Categories", fontsize=30)

plt.plot()
```

## What are the top selling main sub-categories in each of the top 3 categories?

#### 1. Clothes & Accessories

```python
print(all_["Category Level 4"][all_["Meta Category Name"]=="Clothes Shoes & Accessories"].value_counts().head())
print("\nNumber of sub-categories: %d"%all_["Category Level 4"][all_["Meta Category Name"]=="Clothes Shoes & Accessories"].describe()["unique"])
plt.figure(figsize=(30,10))
plt.xticks(fontsize=20)
plt.title("Top 50 Sub-Category", fontsize=30)
all_["Category Level 4"][all_["Meta Category Name"]=="Clothes Shoes & Accessories"].value_counts()[0:50].plot(kind="bar")
plt.plot()
```

##### Important Note:

Temperature in UK in February was 7°C according to http://www.holiday-weather.com/london/averages/february/ <br>
Temperature in UK in March was 9°C according to http://www.holiday-weather.com/london/averages/march/

From February to March, Number of sales of boots dropped from 1066 to 789 <br> while number of sales of coats and jackets dropped from 1019 to 718 in March

#### 3. Computers/Tablets & Networking

```python
print(all_["Category Level 4"][all_["Meta Category Name"]=="Computers/Tablets & Networking"].value_counts().head())
print("\nNumber of sub-categories: %d"%all_["Category Level 4"][all_["Meta Category Name"]=="Computers/Tablets & Networking"].describe()["unique"])
plt.figure(figsize=(30,10))
plt.xticks(fontsize=20)
plt.title("Top 50 Sub-Category", fontsize=30)
all_["Category Level 4"][all_["Meta Category Name"]=="Computers/Tablets & Networking"].value_counts()[0:50].plot(kind="bar")
plt.plot()
```
